```python
from matplotlib import pyplot as plt
import numpy as np
import warnings
import json
import umap

```

```python
warnings.filterwarnings('ignore')
with open('/active/myler_p/People/Sur/J-IP/LtaP/ivan-pacbio/chip_ipd_only.json') as infile:
    data = json.load(infile)
    
vectors = np.array(data['vectors'])[:40000]
labels = np.array(data['labels'])[:40000]
```

```python
fit = umap.UMAP()
embedding = fit.fit_transform(vectors)
positives = (embedding[labels == 1][:,0], embedding[labels == 1][:,1])
negatives = (embedding[labels == 0][:,0], embedding[labels == 0][:,1])
```

```python
plt.figure(dpi = 150)
plt.scatter(negatives[0], negatives[1], label = 'Negatives', s = 1, alpha = 0.1, c = 'C1')
plt.scatter(positives[0], positives[1], label = 'Positives', s = 1, alpha = 0.1, c = 'C0')
plt.axis('off')
plt.title('UMAP Embedding')
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', markerscale = 10)
legend.legendHandles[0].set_alpha(1)
legend.legendHandles[1].set_alpha(1)
plt.tight_layout()
plt.show()
```

```python
vectors = data['vectors']
labels = data['labels']
ipds = vectors[:,200:]
sequences = vectors[:,:200]
```

```python
# Points
n = 1000
# Spacing
t = 1/100
x = np.linspace(0.0, n*t, n)
y = np.sin(2*2*np.pi*x) + 0.5*np.sin(4*2*np.pi*x)
yf = np.abs(np.fft.rfft(y)) * 2/n
xf = np.linspace(0, 1/(2*t), n/2 + 1)

plt.figure(figsize = (6,4), dpi = 100)
plt.plot(xf, yf)
plt.xlim([0, 10])
plt.show()
```

```python
titles = ['A', 'T', 'C', 'G', 'Top IPD', 'Bottom IPD']
for i in range(6):
    total = []
    start = i*50
    stop = start + 50
    for sequence in vectors[classes == 'tp']:
        y = sequence[start:stop]
        yf = np.abs(np.fft.rfft(y)) * 2/50
        total.append(yf)
        
    meanf = np.mean(total, axis = 0)
    xf = np.linspace(0, 1/2, 50/2 + 1)

    plt.figure(figsize = (6,4), dpi = 100)
    plt.plot(xf, meanf)
    plt.title(f'FFT for {titles[i]}')
    if i < 4:
        plt.ylim([0,0.5])
    else:
        plt.ylim([0,2])
    plt.ylabel('Amplitude')
    plt.xlabel('Frequency')
    plt.show()
    plt.close()
```
