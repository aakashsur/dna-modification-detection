```python
from matplotlib import pyplot as plt
from tensorflow import keras
import numpy as np
import umap

import warnings
warnings.filterwarnings('ignore')

data = np.load('/active/myler_p/People/Sur/J-IP/LtaP/ivan-pacbio/data_0.npy')
labels = np.load('/active/myler_p/People/Sur/J-IP/LtaP/ivan-pacbio/labels_0.npy')

x_test = data[-5000:]
y_test = labels[-5000:]
```

```python
sequences = x_test[:,:-102]
ipd = x_test[:,-102:]

print (sequences.shape, ipd.shape)
```

```python
for i in (2, 5, 10, 20, 50, 100, 200):
    fit = umap.UMAP(n_neighbors = i)
    embedding = fit.fit_transform(ipd)
    plt.figure(figsize = (3,3), dpi = 150)
    plt.scatter(embedding[:, 0], embedding[:, 1], c = y_test, s = 1)
    plt.title(f'{i} Neighbors')
    plt.show()

for i in (0.0, 0.1, 0.25, 0.5, 0.8, 0.99):
    fit = umap.UMAP(min_dist = i)
    embedding = fit.fit_transform(ipd)
    plt.figure(figsize = (3,3), dpi = 150)
    plt.scatter(embedding[:, 0], embedding[:, 1], c = y_test, s = 1)
    plt.title(f'{i} Min Distance')
    plt.show()
```
